# Installation des packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Extraction des données

```python
df_1 = pd.read_csv('./1_Warehouse/tests_1.csv')
df_1.head()
```

```python
df_10 = pd.read_csv('./10_Warehouses/tests_10.csv')
df_10.head()
```

```python
df_20 = pd.read_csv('./20_Warehouses/tests_20.csv')
df_20.head()
```

```python
df_30 = pd.read_csv('./30_Warehouses/tests_30.csv')
df_30.head()
```

```python
df_40 = pd.read_csv('./40_Warehouses/tests_40.csv')
df_40.head()
```

```python
df = df_1.join(df_10)
df = df.join(df_20)
df = df.join(df_30)
df = df.join(df_40)
df.head()
```

# Visualisation des données

```python
plt.figure(1, figsize=(20, 20))
fig = plt.gcf()

n_groups = 5

C_mean_normal = df[[
    "C_1_normal", "C_10_normal", "C_20_normal", "C_30_normal", "C_40_normal"
]].mean().values
C_mean_semi = df[[
    "C_1_semi", "C_10_semi", "C_20_semi", "C_30_semi", "C_40_semi"
]].mean().values
C_mean_denormal = df[[
    "C_1_denormal", "C_10_denormal", "C_20_denormal", "C_30_denormal",
    "C_40_denormal"
]].mean().values

C_std_normal = df[[
    "C_1_normal", "C_10_normal", "C_20_normal", "C_30_normal", "C_40_normal"
]].std().values
C_std_semi = df[[
    "C_1_semi", "C_10_semi", "C_20_semi", "C_30_semi", "C_40_semi"
]].std().values
C_std_denormal = df[[
    "C_1_denormal", "C_10_denormal", "C_20_denormal", "C_30_denormal",
    "C_40_denormal"
]].std().values

ax = plt.subplot(221)

index = np.arange(n_groups)
bar_width = 0.3

opacity = 0.4
error_config = {"ecolor": "dimgray"}

rects1 = ax.bar(
    index,
    C_mean_normal,
    bar_width,
    alpha=opacity,
    color="deepskyblue",
    yerr=C_std_normal,
    error_kw=error_config,
    label="normalisation")

rects2 = ax.bar(
    index + bar_width,
    C_mean_semi,
    bar_width,
    alpha=opacity,
    color="lightskyblue",
    yerr=C_std_semi,
    error_kw=error_config,
    label="semi-dénormalisation")

rects3 = ax.bar(
    index + bar_width + bar_width,
    C_mean_denormal,
    bar_width,
    alpha=opacity,
    color="lightblue",
    yerr=C_std_denormal,
    error_kw=error_config,
    label="dénormalisation")

ax.set_xlabel("Nombre des entrepôts")
ax.set_ylabel("Temps moyen d’exécution")
ax.set_title("Requête C")
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(("1_Warehouse", "10_Warehouses", "20_Warehouses",
                    "30_Warehouses", "40_Warehouses"))
ax.legend()

fig.tight_layout()

R_mean_normal = df[[
    "R_1_normal", "R_10_normal", "R_20_normal", "R_30_normal", "R_40_normal"
]].mean().values
R_mean_semi = df[[
    "R_1_semi", "R_10_semi", "R_20_semi", "R_30_semi", "R_40_semi"
]].mean().values
R_mean_denormal = df[[
    "R_1_denormal", "R_10_denormal", "R_20_denormal", "R_30_denormal",
    "R_40_denormal"
]].mean().values

R_std_normal = df[[
    "R_1_normal", "R_10_normal", "R_20_normal", "R_30_normal", "R_40_normal"
]].std().values
R_std_semi = df[[
    "R_1_semi", "R_10_semi", "R_20_semi", "R_30_semi", "R_40_semi"
]].std().values
R_std_denormal = df[[
    "R_1_denormal", "R_10_denormal", "R_20_denormal", "R_30_denormal",
    "R_40_denormal"
]].std().values

ax = plt.subplot(222)

index = np.arange(n_groups)
bar_width = 0.3

opacity = 0.4
error_config = {"ecolor": "dimgray"}

rects1 = ax.bar(
    index,
    R_mean_normal,
    bar_width,
    alpha=opacity,
    color="deeppink",
    yerr=R_std_normal,
    error_kw=error_config,
    label="normalisation")

rects2 = ax.bar(
    index + bar_width,
    R_mean_semi,
    bar_width,
    alpha=opacity,
    color="hotpink",
    yerr=R_std_semi,
    error_kw=error_config,
    label="semi-dénormalisation")

rects3 = ax.bar(
    index + bar_width + bar_width,
    R_mean_denormal,
    bar_width,
    alpha=opacity,
    color="pink",
    yerr=R_std_denormal,
    error_kw=error_config,
    label="dénormalisation")

ax.set_xlabel("Nombre des entrepôts")
ax.set_ylabel("Temps moyen d’exécution")
ax.set_title("Requête R")
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(("1_Warehouse", "10_Warehouses", "20_Warehouses",
                    "30_Warehouses", "40_Warehouses"))
ax.legend()

fig.tight_layout()

U_mean_normal = df[[
    "U_1_normal", "U_10_normal", "U_20_normal", "U_30_normal", "U_40_normal"
]].mean().values
U_mean_semi = df[[
    "U_1_semi", "U_10_semi", "U_20_semi", "U_30_semi", "U_40_semi"
]].mean().values
U_mean_denormal = df[[
    "U_1_denormal", "U_10_denormal", "U_20_denormal", "U_30_denormal",
    "U_40_denormal"
]].mean().values

U_std_normal = df[[
    "U_1_normal", "U_10_normal", "U_20_normal", "U_30_normal", "U_40_normal"
]].std().values
U_std_semi = df[[
    "U_1_semi", "U_10_semi", "U_20_semi", "U_30_semi", "U_40_semi"
]].std().values
U_std_denormal = df[[
    "U_1_denormal", "U_10_denormal", "U_20_denormal", "U_30_denormal",
    "U_40_denormal"
]].std().values

ax = plt.subplot(223)

index = np.arange(n_groups)
bar_width = 0.3

opacity = 0.4
error_config = {"ecolor": "dimgray"}

rects1 = ax.bar(
    index,
    U_mean_normal,
    bar_width,
    alpha=opacity,
    color="orangered",
    yerr=U_std_normal,
    error_kw=error_config,
    label="normalisation")

rects2 = ax.bar(
    index + bar_width,
    U_mean_semi,
    bar_width,
    alpha=opacity,
    color="darkorange",
    yerr=U_std_semi,
    error_kw=error_config,
    label="semi-dénormalisation")

rects3 = ax.bar(
    index + bar_width + bar_width,
    U_mean_denormal,
    bar_width,
    alpha=opacity,
    color="orange",
    yerr=U_std_denormal,
    error_kw=error_config,
    label="dénormalisation")

ax.set_xlabel("Nombre des entrepôts")
ax.set_ylabel("Temps moyen d’exécution")
ax.set_title("Requête U")
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(("1_Warehouse", "10_Warehouses", "20_Warehouses",
                    "30_Warehouses", "40_Warehouses"))
ax.legend()

fig.tight_layout()

D_mean_normal = df[[
    "D_1_normal", "D_10_normal", "D_20_normal", "D_30_normal", "D_40_normal"
]].mean().values
D_mean_semi = df[[
    "D_1_semi", "D_10_semi", "D_20_semi", "D_30_semi", "D_40_semi"
]].mean().values
D_mean_denormal = df[[
    "D_1_denormal", "D_10_denormal", "D_20_denormal", "D_30_denormal",
    "D_40_denormal"
]].mean().values

D_std_normal = df[[
    "D_1_normal", "D_10_normal", "D_20_normal", "D_30_normal", "D_40_normal"
]].std().values
D_std_semi = df[[
    "D_1_semi", "D_10_semi", "D_20_semi", "D_30_semi", "D_40_semi"
]].std().values
D_std_denormal = df[[
    "D_1_denormal", "D_10_denormal", "D_20_denormal", "D_30_denormal",
    "D_40_denormal"
]].std().values

ax = plt.subplot(224)

index = np.arange(n_groups)
bar_width = 0.3

opacity = 0.4
error_config = {"ecolor": "dimgray"}

rects1 = ax.bar(
    index,
    D_mean_normal,
    bar_width,
    alpha=opacity,
    color="limegreen",
    yerr=D_std_normal,
    error_kw=error_config,
    label="normalisation")

rects2 = ax.bar(
    index + bar_width,
    D_mean_semi,
    bar_width,
    alpha=opacity,
    color="lime",
    yerr=D_std_semi,
    error_kw=error_config,
    label="semi-dénormalisation")

rects3 = ax.bar(
    index + bar_width + bar_width,
    D_mean_denormal,
    bar_width,
    alpha=opacity,
    color="lightgreen",
    yerr=D_std_denormal,
    error_kw=error_config,
    label="dénormalisation")

ax.set_xlabel("Nombre des entrepôts")
ax.set_ylabel("Temps moyen d’exécution")
ax.set_title("Requête D")
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(("1_Warehouse", "10_Warehouses", "20_Warehouses",
                    "30_Warehouses", "40_Warehouses"))
ax.legend()

fig.tight_layout()

plt.show()
```
